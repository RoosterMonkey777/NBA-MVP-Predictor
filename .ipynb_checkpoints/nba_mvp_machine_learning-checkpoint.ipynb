We will use a metric called "average percision", this is common for ranking 

**Mean Average Precision (MAP)**:

MAP measures the average precision of the top-ranked items across multiple queries or instances.
It is suitable for evaluating models that rank items in a list where the ordering of items matters.
Higher MAP values indicate better performance.

Other error metrics that might be interesting:

Normalized Discounted Cumulative Gain (NDCG):

NDCG measures the effectiveness of a ranked list by considering both the relevance and the position of items in the list.
It compares the quality of the ranked list produced by the model against an ideal or ground truth ranking.
NDCG values range from 0 to 1, with higher values indicating better performance.
Kendall's Tau:

Kendall's Tau measures the correlation between the rankings produced by the model and the ground truth rankings.
It assesses the pairwise agreement between the rankings, regardless of the actual position in the ranking.
Kendall's Tau ranges from -1 to 1, where 1 indicates perfect agreement and -1 indicates perfect disagreement.
Spearman's Rank Correlation Coefficient:

Spearman's Rank Correlation Coefficient is similar to Kendall's Tau but measures the correlation between the actual ranks and the predicted ranks.
It is based on the ranks of the items rather than their absolute values.
Spearman's correlation coefficient ranges from -1 to 1, with 1 indicating perfect monotonic correlation and -1 indicating perfect inverse correlation.
Mean Reciprocal Rank (MRR):

MRR measures the average reciprocal of the rank of the first relevant item in the list.
It is commonly used in information retrieval tasks where the goal is to retrieve relevant documents or items.
Higher MRR values indicate better performance.

```python
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5)
    predicted = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0 
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found+=1
            ps.append(found/seen)
        seen+=1
    return sum(ps) / len(ps)
```

```python
find_ap(combination)
```

We can use something called backtesting to make predictions for most of the years

```python
years = list(range(1991,2022))
```

```python
aps = []
all_predictions = []
for year in years[5:]:
    train = stats[stats["Year"] < year]
    test = stats[stats["Year"] == year]
    reg.fit(train[predictors],train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    all_predictions.append(combination)
    aps.append(find_ap(combination))
    
```

```python
sum(aps) / len(aps)
```

```python
def add_ranks(predictions):
    predictions = predictions.sort_values("predictions", ascending=False)
    predictions["Predicted_Rk"] = list(range(1,predictions.shape[0]+1))
    predictions = predictions.sort_values("Share", ascending=False)
    predictions["Rk"] = list(range(1,predictions.shape[0]+1))
    predictions["Diff"] = (predictions["Rk"] - predictions["Predicted_Rk"])
    return predictions
```

```python
add_ranks(all_predictions[1])
```

Can build a model to test different models and such

```python
def backtest(stats, model, years, predictors):
    aps = []
    all_predictions = []
    for year in years:
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]
        model.fit(train[predictors], train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, aps, all_predictions = backtest(stats, reg, years[5:], predictors)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
def backtest(stats, model, years, predictors):
    aps = []
    all_predictions = []
    for year in years:
        train = stats[stats["Year"] < year].copy()
        test = stats[stats["Year"] == year].copy()
        sc.fit(train[predictors])
        train[predictors] = sc.transform(train[predictors])
        test[predictors] = sc.transform(test[predictors])
        model.fit(train[predictors],train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, aps, all_predictions = backtest(stats, reg, years[28:], predictors)
```
