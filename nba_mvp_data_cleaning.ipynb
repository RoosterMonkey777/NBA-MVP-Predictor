We want to merge the data from **mvps dataset** to the same players and year in **players dataset**, however some of the dataformat differs between the two datasets (some names in players has a * which is not their on the mvps). To avoid error and have them match all characters, we need to remove the *

```python
# to remove the * we will use the .str.replace() method

#pandas uses * as regex (which has a meaning), so we need to make it false for literal *
players["Player"] = players["Player"].str.replace("*", "", regex=False) 
```
