We want to merge the data from **mvps dataset** to the same players and year in **players dataset**, however some of the dataformat differs between the two datasets (some names in players has a * which is not their on the mvps). To avoid error and have them match all characters, we need to remove the *

```python
# to remove the * we will use the .str.replace() method

#pandas uses * as regex (which has a meaning), so we need to make it false for literal *
players["Player"] = players["Player"].str.replace("*", "", regex=False) 
```

```python
players.head(50)
```

Notice how all the * have been removed

**Also notice in the dataset that their are names being repeated many times (because they played for various teams that season), we just want the total stats so we have to take care of that**

```python
players.groupby(["Player", "Year"])
```

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
```

We can iterate through each group and make sure each group only has one row.

We will do this by writing a function.

```python
def single_row(df):
    if df.shape[0] == 1: # if there is only one row just leave it
        return df
    else:
        row = df[df["Tm"] == "TOT"] # TOT doesnt match the mvp data, so we need to replace it with the last team they played for
        row["Tm"] = df.iloc[-1,:]["Tm"] # the team in the last row
        return row        
```

```python
# now we can use the .apply() method to apply the function to each group
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

We can drop two of the index levels that were created as a result of grouping and combining the dataframes. We can do this by dropping the first two levels

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

Can now see duplicate players only show up once with the last team they played for that season

Now we want to merge our two dataframes : mvps and players

```python
combined_df = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined_df.head()
```

Notice how bow all the columns from the mvps dataset is combined with the players dataset, and because there are no values for those players, they are replaced with NaN

```python
combined_df[combined_df["Pts Won"]> 0]
```

Need to fix tha Nan by replacing with 0's

```python
combined_df[["Pts Won", "Pts Max", "Share"]] = combined_df[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined_df.head()
```

### Cleaning the Team Data

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams.head(20)
```

Notice how in this dataset there are string about the division, we need to remove these.

```python
#  filtering the DataFrame teams to exclude rows where the "W" column contains the string "Division"
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(20)
```

Remove the team names with * in them like before

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head(20)
```

```python
teams["Team"].unique() # look at the unique names in the Teams 
```

```python
combined_df["Tm"].unique()
```

```python
unique_teams_sorted = sorted(combined_df["Tm"].unique())
unique_teams_sorted = sorted(teams["Team"].unique())
unique_teams_sorted
```

Notice how in the combined dataframe, we have the short version of the team names

```python
nicknames = {}

with open("nicknames.csv", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines[1:]: # skip first row 
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
        
```

```python
nicknames
```

can use pandas map method to create in the combined dataframe a full team name column

```python
# applies a mapping function to each value in the "Tm" column. 
# The nicknames variable refers to a mapping function or a dictionary-like object containing mappings.
combined_df["Team"] = combined_df["Tm"].map(nicknames)
```

```python
combined_df.head(10)
```

Now we have everthing to make our final data set with the player stats and the teams names. 

```python
stats = combined_df.merge(teams, how="outer", on=["Team", "Year"])
```
